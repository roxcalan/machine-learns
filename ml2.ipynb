```python
import pandas as pd
data= pd.read_csv("/content/Placement_Data_Full_Class.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['gender']=le.fit_transform(data['gender'])
data['hsc_s']=le.fit_transform(data['hsc_s'])
data['degree_t']=le.fit_transform(data['degree_t'])
data['workex']=le.fit_transform(data['workex'])
data['specialisation']=le.fit_transform(data['specialisation'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X=data.drop(columns=['status'])
y=data['status']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)

classifier = DecisionTreeClassifier(max_depth=5, criterion='gini')

classifier.fit(X_train, y_train)

y_pred=classifier.predict(X_test)

```
